## The Goal
Create a beam of invisible light. Triggers an alarm if someone breaks the beam.

## The Concept: 38kHz Carrier Wave

The sun puts out a lot of Infrared light. So why doesn't your TV remote get confused by the window?

Because TV remotes don't just shine light. They **Flash** it 38,000 times a second (38kHz).
The receiver has sunglasses that block everything EXCEPT 38kHz flashing light.

*   **Transmitter:** Must blink at 38kHz (PWM).
*   **Receiver:** Sees signal (0) or No Signal (1).

## Wiring It Up

We need two modules facing each other.

| Module | Pin | Where it goes on Pico |
| :--- | :--- | :--- |
| **Transmitter** (LED) | **S** | **GP20** (PWM) |
| **Receiver** (Black chip) | **S** | **GP21** (Input) |
| **Receiver** | **v/G** | **3V3** / **GND** |

```python
import machine
import time

# 1. FIRE THE BEAM (PWM 38kHz)
gun = machine.PWM(machine.Pin(20))
gun.freq(38000)
gun.duty_u16(32768) # 50% Power

# 2. Setup Receiver
target = machine.Pin(21, machine.Pin.IN)
led = machine.Pin("LED", machine.Pin.OUT)

while True:
    # If Receiver sees the beam, it touches Ground (0).
    # If Beam is blocked, it floats up (1)
    if target.value() == 1:
        print("ALARM! BEAM BROKEN!")
        led.value(1)
    else:
        led.value(0)
        
    time.sleep(0.1)
```
