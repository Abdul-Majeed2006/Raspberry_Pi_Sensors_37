## The Goal
Detect dark lines on a surface using infrared reflection.

## The Concept: IR Reflection

The **Tracking Sensor** (KY-033) uses an Infrared LED and a receiver.

*   **Reflection:** Light bounces off white/shiny surfaces back to the receiver.
*   **Absorption:** Dark/Black surfaces absorb the light, so nothing bounces back.
*   By checking if we see the beam, we know if we are on a "Line" or "Road".

## Wiring It Up

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP16** |
| **+** (Power) | **3V3** |
| **-** (GND) | **GND** |

```python
import machine
import time

sensor = machine.Pin(16, machine.Pin.IN)
led = machine.Pin("LED", machine.Pin.OUT)

while True:
    # 0 = Reflection Absorbed (On a Black Line)
    # 1 = Reflection Received (On White Surface)
    if sensor.value() == 0:
        print("== ON LINE ==")
        led.value(1)
    else:
        led.value(0)
    
    time.sleep(0.05)
```
