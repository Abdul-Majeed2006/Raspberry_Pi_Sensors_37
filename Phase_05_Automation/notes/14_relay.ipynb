# Lesson 14: High Power (Relays) 🔌

The Pico operates at **3.3 Volts**. That is barely enough to light an LED.
What if you want to turn on a Desk Lamp (120V) or a Fan (12V)?

You need a **Relay**.

## The Goal
Make a mechanical "Click" to switch a separate circuit on and off.

## The Concept: Electromagnetic Switching

Inside the blue box is a shiny coil of copper wire and a metal switch.
1.  We send power to the coil.
2.  The coil becomes a magnet.
3.  The magnet physically pulls the metal switch closed.
4.  *CLICK!*

This keeps our Pico safe because there is NO electrical connection between the Pico and the High Power side. They are isolated.

## Wiring It Up

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP16** |
| **+** (Power) | **3V3** (Use 5V if it doesn't click!) |
| **-** (GND) | **GND** |

```python
import machine
import time

relay = machine.Pin(16, machine.Pin.OUT)

print("Listen for the click...")

while True:
    print("ON")
    relay.value(1)
    time.sleep(2)
    
    print("OFF")
    relay.value(0)
    time.sleep(2)
```
