# Lesson 16: Robot Vision 🤖

How does a Roomba know not to hit the wall? It has eyes.

## The Goal
Create an "Obstacle Avoidance" system.

## The Concept: Reflection

This sensor has two eyes:
1.  **Transmitter:** Shines IR light forward.
2.  **Receiver:** Looks for that light to bounce back.

*   **No Wall:** Light goes on forever. Receiver sees nothing (1).
*   **Wall:** Light bounces back. Receiver sees it (0).

## Wiring It Up

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **OUT** (Signal) | **GP16** |
| **+** (Power) | **3V3** |
| **G** (GND) | **GND** |

```python
import machine
import time

eye = machine.Pin(16, machine.Pin.IN)

while True:
    # 0 means Reflection (Obstacle Detected)
    # 1 means Clear Path
    if eye.value() == 0:
        print("WALL DETECTED! STOP!")
    else:
        print("Path Clear...")
        
    time.sleep(0.2)
```
