# Lesson 12: Photoresistors (Light Sensing) 💡

How do streetlights know when to turn on? They don't have clocks. They have Eyes.

## The Goal
Make an automatic night-light that turns on when the room gets dark.

## The Concept: Photo-Resistance

A **Photoresistor** (or LDR) is a resistor made of Cadmium Sulfide. 
*   **Lots of Light:** Electrons get excited and flow easily (Low Resistance).
*   **Darkness:** Electrons get lazy and block flow (High Resistance).

By reading this with our ADC, we can get a number representing "Brightness".

## Wiring It Up

The photoresistor module usually has 3 pins:

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP26** (ADC0) |
| **+** (Power) | **3V3** |
| **-** (GND) | **GND** |

```python
import machine
import time

light_sensor = machine.ADC(machine.Pin(26))
led = machine.Pin("LED", machine.Pin.OUT)

while True:
    value = light_sensor.read_u16()
    print(f"Light Level: {value}")
    
    # Calibrate this number! Cover the sensor with your hand to see the 'Dark' value.
    if value < 20000: 
        led.value(1)
    else:
        led.value(0)
        
    time.sleep(0.5)
```
