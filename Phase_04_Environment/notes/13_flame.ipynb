# Lesson 13: Fire Detection 🔥

This is a safety device.

## The Goal
Detect an open flame (like a lighter) and sound an alarm.

## The Concept: Infrared Physics

Fire emits visible light (Orange/Yellow), but it emits even MORE **Infrared (IR)** light, which is invisible to human eyes.

This sensor is basically an "Eye" that only sees Infrared. 
*   **Sunlight:** Contains some IR (may cause false alarms).
*   **LEDs:** Contain almost zero IR.
*   **Fire:** Massive amounts of IR.

## Wiring It Up (Clustered)

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **DO** (Digital Out) | **GP16** (Pin 21) |
| **AO** (Analog Out) | **GP26** (Analog 0) |
| **+** (Power) | **3V3** |
| **G** (GND) | **GND** |
| **Buzzer** | **GP17** (Pin 22) |

```python
import machine
import time

# Setup pins in the Bottom Right Cluster
flame = machine.Pin(16, machine.Pin.IN)
buzzer = machine.PWM(machine.Pin(17))

while True:
    # Fire Sensor usually sends 0 when fire is found
    if flame.value() == 0:
        print("FIRE DETECTED!!")
        buzzer.freq(1000)
        buzzer.duty_u16(32768)
    else:
        print("Safe.")
        buzzer.duty_u16(0)
        
    time.sleep(0.5)
```
