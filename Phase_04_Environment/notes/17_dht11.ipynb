## The Goal
Measure both Temperature and Humidity using the digital DHT11 sensor.

## The Concept: Digital Protocols

Unlike the Analog Thermistor, the **DHT11** is a "Smart" sensor. 

*   It has its own tiny processor inside.
*   It sends data as a series of pulses (0s and 1s) over a single wire.
*   This is more accurate than analog for long distances, but requires a special library to read.

## Wiring It Up

We are using the **Digital Temp/Humidity** module (KY-015).

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP16** |
| **Middle** (+) | **3V3** |
| **-** (GND) | **GND** |

```python
import machine
import dht
import time

sensor = dht.DHT11(machine.Pin(16))

while True:
    try:
        sensor.measure()
        t = sensor.temperature()
        h = sensor.humidity()
        
        print(f"Temp: {t}Â°C | Humidity: {h}%")
        
    except OSError as e:
        print("Sensor failure!")
        
    time.sleep(2) # DHT11 needs time to breathe!
```
