Let's see your pulse!

## The Goal
Detect your heartbeat using light absorption.

## The Concept: Photoplethysmogram (PPG)

That's a big word for a simple idea: 

*   Blood is thick and absorbs light.
*   Every time your heart beats, a surge of blood goes to your fingertips.
*   The **Heartbeat Sensor** (KY-039) shines an IR light through your finger.
*   The detector on the other side sees LESS light during the "pump" phase of your heart.
*   By watching the peaks and valleys, we can calculate your BPM (Beats Per Minute).

## Wiring It Up

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP26** (ADC0) |
| **Middle** (+) | **3V3** |
| **-** (GND) | **GND** |

> **‚ö†Ô∏è NOTE:** This sensor is VERY sensitive. You must keep your finger perfectly steady.

```python
import machine
import time

sensor = machine.ADC(machine.Pin(26))

print("Place finger gently... Reading heart data...")

while True:
    # Read the raw light level
    val = sensor.read_u16()
    
    # Print to the 'Plotter' (Output console)
    # You should see high numbers dip when blood surges!
    print(val)
    
    time.sleep(0.02)
```
