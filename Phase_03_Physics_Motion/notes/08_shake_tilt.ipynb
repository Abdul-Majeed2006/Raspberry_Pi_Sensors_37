# Lesson 08: Shake & Tilt 🎿

In this lesson, we give our Pico the sense of **Balance** and **Impact**.

## The Goal
Detect if the device is being shaken or knocked.

## The Concept: Mechanical Physics

### 1. Vibration Switch (Shock Sensor)
Inside this module is a tiny **Spring** surrounding a metal pin. 
When you shake it, the spring vibrates and touches the pin, closing the circuit for a split second.

### 2. Tilt Switch (Ball Switch)
Similar to the Light Cup, this has a metal ball. 
*   Tilt Up = Ball rolls onto contacts (ON).
*   Tilt Down = Ball rolls away (OFF).

## Wiring It Up

Both sensors work exactly the same way: as a switch.

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP16** |
| **+** (Power) | **3V3** |
| **-** (GND) | **GND** |

```python
import machine
import time

# Setup Input with PULL_UP
# This means the pin connects to 3.3V inside the Pico.
# When the sensor triggers, it connects to Ground (0).
sensor = machine.Pin(16, machine.Pin.IN, machine.Pin.PULL_UP)
led = machine.Pin("LED", machine.Pin.OUT)

print("Waiting for impact...")

while True:
    # If sensor connects (0), we detected a shake
    if sensor.value() == 0:
        print("Shake!")
        led.value(1)
        time.sleep(0.1)
        led.value(0)
```
