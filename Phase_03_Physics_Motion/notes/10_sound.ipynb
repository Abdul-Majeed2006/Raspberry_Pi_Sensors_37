# Lesson 10: Sound Detection 🎤

This sensor allows the Pico to hear the world. 

## The Goal
Detect a loud noise (like a Clap) and toggle a light.

## The Concept: Sound Thresholds

The microphone module listens to sound waves.
It has a small **blue box with a screw** on it (Potentiometer).

*   **Turn Clockwise:** More Sensitive (Triggers on whispers).
*   **Turn Counter-Clockwise:** Less Sensitive (Triggers only on explosions).

You will need to use a small screwdriver to tune it until the little light on the module flickers when you clap.

## Wiring It Up

| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **OUT** (Signal) | **GP16** |
| **+** (Power) | **3V3** |
| **G** (GND) | **GND** |

```python
import machine
import time

# Setup Mic
mic = machine.Pin(16, machine.Pin.IN)

print("Listening...")

while True:
    if mic.value() == 1: # Detected Sound
        print("LOUD NOISE!")
        time.sleep(0.2) # Wait for silence
```
