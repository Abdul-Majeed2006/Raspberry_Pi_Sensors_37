# Lesson 01: Light It Up (RGB LEDs)

Welcome to your first step in Physical Computing! Today, we are going to learn how to control color using code.

## The Goal
We will make an LED glow in any color we want (Red, Green, Blue, Purple, or even Orange!).

## The Concept: What is an RGB LED?
An **RGB LED** is just a tiny package with three small lights inside it: **Red**, **Green**, and **Blue**.
By changing the brightness of each one, we can mix them to make over 16 million colors.

### PWM (Pulse Width Modulation)
Computers are digital (ON/OFF). To dim an LED, we turn it ON and OFF really fast (thousands of times a second).
- ON mostly = Bright
- OFF mostly = Dim
This concept is called **PWM**.

## Wiring It Up
| Pin on LED Module | Where it goes on Pico |
| :--- | :--- |
| **R** (Red) | **GP13** |
| **G** (Green) | **GP14** |
| **B** (Blue) | **GP15** |
| **-** (GND) | **Any GND pin** |

```python
import machine
import time

# 1. Setup the Pins
red_led = machine.PWM(machine.Pin(13))
green_led = machine.PWM(machine.Pin(14))
blue_led = machine.PWM(machine.Pin(15))

# 2. Set the Speed (Frequency)
red_led.freq(1000)
green_led.freq(1000)
blue_led.freq(1000)

def set_color(r, g, b):
    red_led.duty_u16(r)
    green_led.duty_u16(g)
    blue_led.duty_u16(b)

print("Starting Color Cycle!")

while True:
    print("Red!")
    set_color(65535, 0, 0)
    time.sleep(1)

    print("Green!")
    set_color(0, 65535, 0)
    time.sleep(1)

    print("Blue!")
    set_color(0, 0, 65535)
    time.sleep(1)
```
