# Lesson 02: The Magic Light Cup

In this lesson, we are dealing with **Position & Physics**.

## The Goal
We want the Pico to know if it is being tilted upside down. When it tilts, we change the LED brightness.

## The Concept: The Motion Switch
Inside the module is a tiny ball (or mercury in some models) that rolls with gravity.
- **Level:** Connects the circuit (ON).
- **Tilted:** Breaks the circuit (OFF).
It is a basic mechanical orientation sensor.

## Wiring It Up (Clustered)
| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP16** (Pin 21) |
| **L** (LED) | **GP17** (Pin 22) |
| **+** (VCC) | **3V3** |
| **G** (GND) | **GND** |

```python
import machine
import time

tilt_switch = machine.Pin(16, machine.Pin.IN)
led = machine.PWM(machine.Pin(17))
led.freq(1000)

print("System Active. Tilt the sensor!")

while True:
    status = tilt_switch.value()

    if status == 1:
        print("Connected!")
        led.duty_u16(65535)
    else:
        print("Broken Circuit")
        led.duty_u16(5000)
        
    time.sleep(0.1)
```
