# Lesson 02: The Magic Light Cup

In this lesson, we are dealing with **Position & Physics**.

## The Goal
We want the Pico to know if it is being tilted upside down. When it tilts, we change the LED brightness.

## The Concept: The Mercury Switch
Inside the glass bulb is a tiny ball of **Mercury** (liquid metal).
- **Upright:** Connects the circuit (ON).
- **Tilted:** Breaks the circuit (OFF).
It is a switch controlled by gravity.

## Wiring It Up
| Pin on Module | Where it goes on Pico |
| :--- | :--- |
| **S** (Signal) | **GP16** |
| **L** (LED) | **GP15** |
| **+** | **3V3** |
| **G** | **GND** |

```python
import machine
import time

tilt_switch = machine.Pin(16, machine.Pin.IN)
led = machine.PWM(machine.Pin(15))
led.freq(1000)

print("Tilt the sensor!")

while True:
    status = tilt_switch.value()

    if status == 1:
        print("Connected!")
        led.duty_u16(65535)
    else:
        print("Broken Circuit")
        led.duty_u16(2000)
        
    time.sleep(0.1)
```
